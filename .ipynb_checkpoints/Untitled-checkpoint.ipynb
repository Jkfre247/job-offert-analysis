```python
import nltk
import pandas as pd

# Pobranie zbioru danych testowych (przykładowe opisy ofert pracy)
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')


```

```python
# Przykładowy DataFrame
df = pd.DataFrame({'Description': ['We are looking for a software developer with at least 3 years of experience in Python and Java. The candidate should have experience in machine learning and natural language processing.',
                                   'We need a project manager with at least 5 years of experience in software development. The candidate should be able to manage a team and work under tight deadlines.']})

# Funkcja do rozpoznawania nazwanych jednostek
def named_entity_recognition(text):
    sentences = nltk.sent_tokenize(text)
    tokenized_sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
    tagged_sentences = [nltk.pos_tag(sentence) for sentence in tokenized_sentences]
    chunked_sentences = nltk.ne_chunk_sents(tagged_sentences, binary=False)
    return chunked_sentences

# Zastosowanie funkcji do kolumny "Description" w DataFrame
df['Named Entities'] = df['Description'].apply(named_entity_recognition)

# Wyświetlenie DataFrame z kolumną "Named Entities"
print(df['Named Entities'])
for entities in df['Named Entities']:
    for sentence in entities:
        for entity in sentence:
            if hasattr(entity, 'label'):
                print(entity.label(), ' '.join(word for word, tag in entity.leaves()))
```

```python
import matplotlib.pyplot as plt

# Wyodrębnienie jednostek ze wszystkich opisów ofert pracy
all_entities = []
for entities in df['Named Entities']:
    for entity in entities:
        all_entities.extend([ne for ne in entity if isinstance(ne, nltk.tree.Tree)])

# Policzenie wystąpień każdej jednostki
entity_counts = {}
for entity in all_entities:
    if entity.label() in entity_counts:
        entity_counts[entity.label()] += 1
    else:
        entity_counts[entity.label()] = 1

# Wyświetlenie wykresu kołowego z ilością wystąpień poszczególnych jednostek
labels = list(entity_counts.keys())
counts = list(entity_counts.values())
plt.pie(counts, labels=labels, autopct='%1.1f%%')
plt.title('Named Entity Recognition')
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Utwórz listę kluczowych fraz
key_phrases_list = df['Key Phrases'].tolist()
flat_list = [item for sublist in key_phrases_list for item in sublist]

# Połącz kluczowe frazy w jeden ciąg tekstowy
text = ' '.join(flat_list)

# Wygeneruj wordcloud
wordcloud = WordCloud(width=800, height=800, background_color='white', min_font_size=10).generate(text)

# Wyświetl wordcloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
