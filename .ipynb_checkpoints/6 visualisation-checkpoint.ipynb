```python
programming_languages = ['tensorflow', 'kafka',"apache", "excel","panda","tableau","kera", 'azure', 'google cloud', 'hadoop', 'microsoft', 'python', 'scala', 'spark', 'hive', 'sql', 'shiny', 'kera', 'numpy', 'scipy', 'pandas', 'matlab', 'nifi', 'streamsets', 'sphinx', 'nutch', 'solr', 'jupyter', 'qlik', 'powerbi', 'postgresql', 'mongodb', 'cassandra', 'soap', 'rest', 'mahout', 'accumulo', 'spss', 'modeler', 'redis', 'elastic', 'powershell', 'lynx', 'oracle', 'teradata', 'api', 'json',  'soup', 'scrapy', 'nosql', 'mongodb', 'celery', 'docker', 'kubernetes', 'pytorch', 'caffe', 'theano', 'hbase', 'spacy', 'apache', 'opencv', 'scikit', 'raspberry', 'amqp', 'rabbitmq', 'snowflake', 'digitalocean', 'gurobi', 'hexagon', 'console', 'crimson', 'beam']

cloud_services = ['azure', 'google',"cloud", 'aws', 'databricks', 'snowflake', 'elastic', 'mapreduce', 'redshift', 'glue',"amazon" ,'kong', 'digitalocean']

big_data_tools = ['hadoop', 'spark', 'hive', 'mahout', 'accumulo', 'nifi', 'streamsets', 'sphinx', 'nutch', 'solr',"sql", 'postgresql', 'mongodb', 'cassandra', 'redis', 'elastic', 'search', 'hbase', 'spacy', 'apache', 'opencv', 'scikit', 'beam']

machine_learning_frameworks = ['tensorflow', 'kera', 'pytorch', 'theano', "powerbi"]

business_intelligence_tools = ['azure', 'google cloud', 'powerbi', 'qlik', 'tableau', 'pentaho', 'birt', 'obiee']

programming_languages.extend(cloud_services)
programming_languages.extend(big_data_tools)
programming_languages.extend(machine_learning_frameworks)
programming_languages.extend(business_intelligence_tools)

print(programming_languages)
```

removes unnecessary words and scans for relevant technology-related words

```python
programingl=[]
for i in entities2[entities2["Category"] == "Product"]["Entity"]:
    x=i.split(" ")
    if len(x)>1:
        for j in x:
            if j in programming_languages:
                programingl.append(j)
    else:
        programingl.append(i)
for i in entities2[entities2["Category"] == "Skill"]["Entity"]:
    x=i.split(" ")
    if len(x)>1:
        for j in x:
            if j in programming_languages:
                programingl.append(j)
    elif i in programming_languages:
        programingl.append(i)
    
tab= ['bike', 'vehicle', 'car', 'book', 'drug', 'cd', 'laptop','food', 'fruit', 'computer',"coffee", "phone",'drink']
programingl = [x for x in programingl if x not in tab]

wordcloud = WordCloud(width=800, height=800, background_color="white", max_words=50).generate(" ".join(programingl))

# plot word cloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()        
```

```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
freq = pd.Series(programingl).value_counts()

def plot_bar_chart(min_count):
    filtered_freq = freq[freq >= min_count]
    plt.figure(figsize=(10, 6))
    plt.bar(filtered_freq.index, filtered_freq.values)
    plt.xticks(rotation=90)
    plt.show()

slider = widgets.IntSlider(value=10, min=8, max=freq.max(), step=1, description='Minimum count:')
widgets.interact(plot_bar_chart, min_count=slider)
```

This part is responsible for finding the number of occurrences of a specific word in the text and then writing this down in order to convert it into a visualisation in the tableau. Unfortunately, I found that the NER did not work as I wanted and removed some words, so I had to choose another option where I searched for words by key. At least the NER helped me with one thing: defining all the technologies.

```python
import re
l=[]
def count_word_occurrences(word):
    descriptions_with_word = set()
    pattern = re.compile(r'\b' + word + r'\W', re.IGNORECASE)
    for description in sprawdzam["Description"]:
        if pattern.search(description):
            descriptions_with_word.add(description)
    return word, len(descriptions_with_word)
for i in programming_language_s:
    try:
        word,count=count_word_occurrences(i)
        l.append({"Word":word,"Category":"Programing Language","Count":count})
    except:
        print(i)
for i in frameworks:
    word,count=count_word_occurrences(i)
    if word=="tensorflow":
        word="tensorf"
    l.append({"Word":word,"Category":"frameworks","Count":count})
for i in bigdata:
    word,count=count_word_occurrences(i)
    l.append({"Word":word,"Category":"bigdata","Count":count})
for i in business_intelligence:
    word,count=count_word_occurrences(i)
    l.append({"Word":word,"Category":"business intelligence","Count":count})
for i in cloud_services:
    word,count=count_word_occurrences(i)
    l.append({"Word":word,"Category":"cloud services","Count":count})
xz = pd.DataFrame(l,columns=["Word","Category","Count"])
xz.to_csv("data2.csv",index=False)
#word = "python"
#count = count_word_occurrences(word)
#print(f"{word} wystpuje {count} razy.")
```
