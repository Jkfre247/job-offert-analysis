## Separation of countries, regions, cities from locations and repair of damaged records


```python
#assign value function
def sepatate_3(df,i, local):
    df.loc[i, "City"] = local[0]
    df.loc[i, "Region"] = local[1]
    df.loc[i, "Country"] = local[2]
    return df.loc[i]

def separate_localisation(df):
    for i in df["Localisation"].unique():
        print(i)
# For Poland         
def separate_localisation_pl(df):
    for i in range(len(df)):
        local = (unidecode(df.loc[i, "Localisation"])).split(", ")
        if len(local) == 3:
            df.loc[i]=sepatate_3(df,i,local)
        elif not (local[0]=="Polska"):
            df.loc[i,"City"]=local[0].split(" i ")[0]
            if "Nowy" in local[0]:
                df.loc[i, "Region"] = "Woj. Malopolskie"
            else:
                df.loc[i, "Region"] = "Woj. Pomorskie"
        df.loc[i, "Country"] = "Polska"
    return df
# for german
def separate_localisation_de(df):
    for i in range(len(df)):
        local = (unidecode(df.loc[i, "Localisation"])).split(", ")
        if len(local) == 3:
            df.loc[i]=sepatate_3(df,i, local)
        elif (len(local)==2):
            df.loc[i,"Country"]=local[1]
            df.loc[i,"City"]=local[0]
        else:
            if re.search(".* i .*",local[0]):
                df.loc[i,"City"]=local[0].split(" i ")[0]
            else:
                df.loc[i,"Country"]=local[0]
    return df

#for uk
def separate_localisation_uk(df):
    for i in range(len(df)):
        local = (unidecode(df.loc[i, "Localisation"])).split(", ")
        if len(local) == 3:
            df.loc[i]=sepatate_3(df,i, local)
        elif (len(local)==2):
            df.loc[i,"Region"]=local[0]
        df.loc[i,"Country"]="Wielka Brytania"
    return df

#print(separate_localisation_uk(jobs_uk)["Country"].unique())   
#separate_localisation(jobs_pl)
# you have to create a special one for Usa because they have a different distribution 
#of 2 city and state data without the name of the country
def separate_localisation_us(df):
    for i in range(len(df)):
        local = (unidecode(df.loc[i, "Localisation"])).split(", ")
        if len(local) == 3:
            df.loc[i]=sepatate_3(df,i, local)
        elif (len(local)==2):
            df.loc[i,"City"]=local[0]
            if not re.search(".*Stany.*",local[1]):
                df.loc[i,"Region"]=local[1]
        df.loc[i,"Country"]="USA"
    return df
```

```python
jobs_pl=separate_localisation_pl(jobs_pl)
jobs_de=separate_localisation_de(jobs_de)
jobs_uk=separate_localisation_uk(jobs_uk)
jobs_us=separate_localisation_us(jobs_us)
```

## repair df

```python
#naprawa krajów baza niemiecka
#jobs_de['Country'] = jobs_de['Country'].where(jobs_de['Country'].isin(['', 'Rejon Stuttgart']), jobs_de['Country'])
jobs_de['Country'] = jobs_de['Country'].where(jobs_de['City'] != 'Bazylea', 'Szwajcaria')
jobs_de['Country'] = jobs_de['Country'].where(jobs_de['Country'] != '', 'Niemcy')
jobs_de['Country'] = jobs_de['Country'].where(jobs_de['Country'] != 'Rejon Stuttgart', 'Niemcy')

jobs_uk.drop(jobs_uk[jobs_uk['Region'] == "Chicago"].index, inplace=True)
jobs_uk.loc[jobs_uk['Region'] == "Londyn i okolice", ['Region', 'City']] = ["Anglia", "London"]
jobs_us.loc[~jobs_us["Level"].isin(["Poczatkujacy", "Staz"]), "Type"] = jobs_us["Level"]
jobs_us.loc[~jobs_us["Level"].isin(["Poczatkujacy", "Staz"]), "Level"] = "Poczatkujacy"
jobs_uk.loc[~jobs_uk["Level"].isin(["Poczatkujacy", "Staz"]), "Type"] = jobs_uk["Level"]
jobs_uk.loc[~jobs_uk["Level"].isin(["Poczatkujacy", "Staz"]), "Level"] = "Poczatkujacy"

```

## merging the data into new dataframes after data science categories, etc.

```python

Data_Scientist=pd.concat([jobs_us.loc[jobs_us["Role"]=="Data Scientist"],jobs_uk.loc[jobs_uk["Role"]=="Data Scientist"],jobs_pl.loc[jobs_pl["Role"]=="Data Scientist"],jobs_de.loc[jobs_de["Role"]=="Data Scientist"]],axis=0)
Data_Analyst=pd.concat([jobs_us.loc[jobs_us["Role"]=="Data Analyst"],jobs_uk.loc[jobs_uk["Role"]=="Data Analyst"],jobs_pl.loc[jobs_pl["Role"]=="Data Analyst"],jobs_de.loc[jobs_de["Role"]=="Data Analyst"]],axis=0)
AI=pd.concat([jobs_us.loc[jobs_us["Role"]=="A.I."],jobs_uk.loc[jobs_uk["Role"]=="A.I."],jobs_pl.loc[jobs_pl["Role"]=="A.I."],jobs_de.loc[jobs_de["Role"]=="A.I."]],axis=0)
ML=pd.concat([jobs_us.loc[jobs_us["Role"]=="Machine Learning"],jobs_uk.loc[jobs_uk["Role"]=="Machine Learning"],jobs_pl.loc[jobs_pl["Role"]=="Machine Learning"],jobs_de.loc[jobs_de["Role"]=="Machine Learning"]],axis=0)
```

# drop dupplicates

```python
AI=AI.drop_duplicates(subset="Description")
ML=ML.drop_duplicates(subset="Description")
Data_Scientist=Data_Scientist.drop_duplicates(subset="Description")
Data_Analyst=Data_Analyst.drop_duplicates(subset="Description")
```
