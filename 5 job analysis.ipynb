## delete special characters
getting rid of unnecessary characters getting rid of the Polish beginning which has not been translated
after completing the project, I noticed that I might not have removed some characters, e.g. - or #, as this eliminated some libraries. However, there were also pitfalls, such as pytho/r/scala, which would not distinguish between individual words

```python
def clean(df):
    rows_clean = []
    for index, row in df.iterrows():
        row=re.sub("[\n]"," ",row['Description'])
        row=re.sub("Ta oferta została pozyskana z tablicy ofert pracy. Więcej informacji ","",row)
        row_clean = re.sub(r"[-—()\"$@;:{}`’+=~|.!?,'0-9]"," ",row)
        rows_clean.append({'Description': row_clean.lower()})
    df_clean = pd.DataFrame(rows_clean, columns=["Description"])
    return df_clean

df_desc=clean(df)

```

# lametisation 
in this stage, it was necessary to assign to each word its corresponding part of speech in order to carry out the lametisation correctly, removing the stopword to improve the analysis 

```python
stop_words = set(nltk.corpus.stopwords.words('english'))
wnl = WordNetLemmatizer()
def lammetize_words(text):
    for word, tag in pos_tag(word_tokenize(text)):
        wntag = tag[0].lower()
        wntag = wntag if wntag in ['a', 'r', 'n', 'v'] else None
        if not wntag:
            lemma = word
        else:
            lemma = wnl.lemmatize(word, wntag)
    return lemma
def pre_process(text):
    text.loc[:, "Description"] = text["Description"].str.strip()
    text.loc[:, 'Description'] = text['Description'].apply(lambda x: ' '.join([ word for word in word_tokenize(x)  if not word in stop_words]))
    #text = text.apply(lambda x: x.strip())          #Rdeleting useless spaces
    #text["Description"] = text["Description"].apply(lambda x: " ".join(x.translate(str.maketrans('', '', string.punctuation)) for x in x.split() if x.isalpha()))
    #text["Description"] = text["Description"].apply(lambda x: " ".join(x for x in x.split() if x not in stop)) #Removing stop words
    #text["Description"] = text["Description"].apply(lambda x: str(TextBlob(x).correct()))                      #Correct spelling corrections
    #text.loc[:, "Description"] = text["Description"].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))   #lemmatization
    text['Description'] = text['Description'].apply(lambda x: ' '.join([ lammetize_words(word) for word in word_tokenize(x)]))
    #text.loc[:, "Description"] = text["Description"].apply(lambda x: " ".join([Word(word).lemmatize(pos_tag([word])[0][1].lower())for word in word_tokenize(x)]))
    return(text)
```

# reduce size
function had the task of reducing the string into smaller pieces because there were limitations in the processing of the data by azure cognitive services

```python
def reduce_size(df, max_words=400):
    desc = []
    #num=0
    for index, row in df.iterrows():
        # Podziel tekst na słowa
        words = row["Description"].split(" ")
        # Podziel słowa na części o długości nie większej niż max_words
        while len(words) > 0:
            #if index==677:
                #print(num)
            part = " ".join(words[:max_words])
            desc.append({"Description": part})
            #num+=1
            words = words[max_words:]
    # Stwórz nowy DataFrame na podstawie listy desc
    new_df = pd.DataFrame(desc, columns=["Description"])
    return new_df
```

```python
import os
import typing
from azure.core.credentials import AzureKeyCredential
from azure.ai.textanalytics import TextAnalyticsClient


def recognise_entities(data):
    
    endpoint = ""
    key = ""
    text_analytics_client = TextAnalyticsClient(endpoint=endpoint, credential=AzureKeyCredential(key))
    df_n=reduce_size(data)
    documents = df_n["Description"].tolist()
    print(len(documents))
    entities_list = []
    for i in range(0, len(documents), 5):
        batch = documents[i:] if i+5 >len(documents) else documents[i:i+5]
        result = text_analytics_client.recognize_entities(batch)
        # assigning a value to a word and its category
        for doc in result:
            for entity in doc.entities:
                entity_dict = {"Entity": entity.text, "Category": entity.category}
                entities_list.append(entity_dict)

    entities_df = pd.DataFrame(entities_list, columns=["Entity", "Category"])
    return entities_df


df_entities = recognise_entities(data)
```

```python
# tests with key phrases however i decided to choose ner instead
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from wordcloud import WordCloud
import matplotlib.pyplot as plt


# Function identify key phrases
def extract_key_phrases(text):
    # Divide text into tokens
    tokens = word_tokenize(text)
    # Extract only nouns and adjectives
    pos_tags = nltk.pos_tag(tokens)
    nouns_adj = [word for word, pos in pos_tags if (pos == 'NN' or pos == 'JJ')]
    # Calculate the frequency of tokens and return the 10 most frequent ones
    fdist = FreqDist(nouns_adj)
    return fdist.most_common(10)

# Add a column of key phrases to the DataFrame

data['Key Phrases'] = data['Description'].apply(extract_key_phrases)

# Convert the list column into separate rows in the DataFrame
key_phrases = pd.DataFrame(data['Key Phrases'].tolist(), index=data.index)

key_phrases_list = [[phrase for phrase, _ in row] for row in data['Key Phrases']]

text = ' '.join([phrase for sublist in key_phrases_list for phrase in sublist])

# word cloud generate
wordcloud = WordCloud(width=800, height=800, background_color='white', min_font_size=10,max_words=50).generate(text)

# Wyświetl wordcloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
